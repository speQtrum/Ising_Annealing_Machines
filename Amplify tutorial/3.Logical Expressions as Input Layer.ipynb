```
from amplify import LogicPoly, gen_symbols, intersection, union, symmetric_difference
```

```
####### Boolean logic  ##########
print('Boolean OR gate:\n',LogicPoly(0) | LogicPoly(1))           # OR gate

print('\nBoolean AND gate:\n',LogicPoly(0) & LogicPoly(1))        # AND gate

print('\nBoolean XOR gate:\n',~(LogicPoly(0) & LogicPoly(1)))     # XOR gate

print('\nBoolean NOR gate:\n',~(LogicPoly(0) | LogicPoly(1)))     # NOR gate

print('\nBoolean NAND gate:\n',~(LogicPoly(0) & LogicPoly(1)))    # NAND gate
```

### **Auxilary function from list**

```
X = gen_symbols(LogicPoly,3)
print(X)
```

```
print('Interaction:\n',intersection(X))

print('\nUnion:\n',union(X))

print('\nSymmetric difference:\n',symmetric_difference(X))
```

### **On the fly auxilary function construction**

```
##########  Using intersection method  ###########3

print('Including all terms:\n',intersection(3))

print('\nAll terms within range:\n', intersection(1,4))

print('\nAll even terms in a range:\n',intersection(0,6,2))

print('\nAll odd terms within an interval with range object:\n',intersection(range(1,6,2)))
```

```
##########  Using Union method  ###########3

print('Including all terms:\n',union(3))

print('\nAll terms within range:\n', union(1,4))

print('\nAll even terms in a range:\n',union(0,6,2))

print('\nAll odd terms within an interval with range object:\n',union(range(1,6,2)))
```

```
##########  Using symmetric difference method  ###########3

print('Including all terms:\n',symmetric_difference(3))

print('\nAll terms within range:\n', symmetric_difference(1,4))

print('\nAll even terms in a range:\n',symmetric_difference(0,6,2))

print('\nAll odd terms within an interval with range object:\n',symmetric_difference(range(1,6,2)))
```

### **Using Logic expression's methods**

```
p = ~LogicPoly(0) | LogicPoly(1)

print(p)
```

```
print(p.to_Poly())  # converting logical expressions to Polynomial
```

```
p.max_index()  # maximum index appearing in the polynomial 
```

```
##### Evaluating the logical expression  ########

print(p.replace({0:3}))  ## replacing 1st variable

print(p.replace_all([3,6]))  ## replacing all variable
```

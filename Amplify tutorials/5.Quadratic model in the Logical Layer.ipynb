```
from amplify import BinaryPoly, BinaryMatrix, BinaryQuadraticModel, gen_symbols, sum_poly
from amplify.constraint import equal_to
```

## **Binary Quadratic model from Binary polynomial**

```
q = gen_symbols(BinaryPoly,3)

f = 2*q[0]*q[1] - 3*q[1]*q[2] + 6*q[0] + 7*q[1] + 8*q[2] + 5

c = 4*equal_to(sum_poly(q),2)  # constraint

model = BinaryQuadraticModel(f,c)

print('Variable array:',q)
print('Binary Polynomial is: \n',f)
print('Constraint:\n',c)
```

```
############# Attributes of Binary Quadratic model  ##################3

print('Input polynomial:\n',model.input_poly)   

print('\nInput matrix:\n',model.input_matrix)

print('\nNumber of input variables:',model.num_input_vars)

print('\nNumber of logical variables:',model.num_logical_vars)

print('\nOver all logical polynomial:\n',model.logical_model_poly)

print('\nOver all logical matrix:\n',model.logical_model_matrix)

print('\nLogical mapping:\n',model.logical_mapping)

print('\nInput constraint:\n',model.input_constraints)

print('\nConstraint satisfaction test:\n',model.check_constraints([1,0,1]))
```
